## Understanding Positive and Negative relationships with different cases 

```python
#import all the necessary libraries
import networkx as nx
import matplotlib.pyplot as plt
import random
```

#### There are 4 basic steps in creating a graph (fully connected or partially connected graph)

**Step-1 :** Create a graph object using networkx package

```python
#creating a graph object
pos_neg_relation_graph = nx.Graph()
```

**Step-2 :**Add nodes or vertices to the graph

```python
#add nodes to the created graph
pos_neg_relation_graph.add_node('0')
pos_neg_relation_graph.add_node('1')
pos_neg_relation_graph.add_node('2')
pos_neg_relation_graph.add_node('3')
pos_neg_relation_graph.add_node('4')
pos_neg_relation_graph.add_node('5')
pos_neg_relation_graph.add_node('6')
```

```python
#see the nodes in the graph
pos_neg_relation_graph.nodes
```

**Step-3 :**Add edges to the graph (which is like adding connections to the graph)

```python
#adding the edges to the graph
pos_neg_relation_graph.add_edge('0','1',color='r',weight=3)  #color parameter is used to assign the colour to the particular edge
pos_neg_relation_graph.add_edge('0','2',color='r',weight=3)  #weight parameter (how strong the connection should be) 
pos_neg_relation_graph.add_edge('0','3',color='g',weight=3)  #visualize the graph to understand how weight parameter is used
pos_neg_relation_graph.add_edge('0','4',color='g',weight=4)
pos_neg_relation_graph.add_edge('0','5',color='g',weight=4)
pos_neg_relation_graph.add_edge('0','6',color='g',weight=4)
```

```python
pos_neg_relation_graph.edges
```

**Step-4**:Visualize the created graph

```python
#draw is method within networkx module
nx.draw(pos_neg_relation_graph,with_labels=1)  #with_labels = True ---> will represent labels in the created nodes
```

### Styling the created graph (labels renaming, assigning colors and weights to the edges)

Relabelling --- Changing the labels of already created graph

```python
#adding node names 
labels_dict = {}
labels_dict['0']='Fedde'
labels_dict['1']='Friend_1'
labels_dict['2']='Friend_2'
labels_dict['3']='Friend_3'
labels_dict['4']='Friend_4'
labels_dict['5']='Friend_5'
labels_dict['6']='Friend_6'
```

Adding colors and weights to already created graph

```python
#adding colors and edges to the created graph
edges = pos_neg_relation_graph.edges()
colors = [pos_neg_relation_graph[u][v]['color'] for u,v in edges] #list comprehension to retrieve the assigned colors to edges 
weights = [pos_neg_relation_graph[u][v]['weight'] for u,v in edges]                                          #     and weights
```

```python
plt.figure(figsize=(6,6))
nx.draw(pos_neg_relation_graph,with_labels=True,labels=labels_dict,edges=edges, edge_color=colors,width=weights,node_size=5000)
```

# STABLE FRIENDSHIPS 
## STABLE relationship

### Case-1 : Only one positive relationship exist (There is one common enemy for two people. So, these two people became friends)

```python
stable_friendship_graph_1 = nx.Graph()
```

```python
#https://stackoverflow.com/questions/28533111/plotting-networkx-graph-with-node-labels-defaulting-to-node-name
stable_friendship_graph_1.add_node('0')
stable_friendship_graph_1.add_node('1')
stable_friendship_graph_1.add_node('2')
```

```python
stable_friendship_graph_1.add_edge('0','1',color='r',weight=3)
stable_friendship_graph_1.add_edge('0','2',color='r',weight=3)
stable_friendship_graph_1.add_edge('1','2',color='g',weight=4)
```

```python
#adding node names 
labels_dict = {}
labels_dict['0']='Anoy'
labels_dict['1']='Cures'
labels_dict['2']='Seret'
```

```python
#https://stackoverflow.com/questions/25639169/networkx-change-color-width-according-to-edge-attributes-inconsistent-result

#adding edges
edges = stable_friendship_graph_1.edges()
colors = [stable_friendship_graph_1[u][v]['color'] for u,v in edges]
weights = [stable_friendship_graph_1[u][v]['weight'] for u,v in edges]
```

```python
plt.figure(figsize=(6,6))
nx.draw(stable_friendship_graph_1,with_labels=True,labels=labels_dict,edges=edges, edge_color=colors,width=weights,node_size=5000)
plt.show()
```
